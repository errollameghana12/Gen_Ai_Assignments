```python
# Job Description
job_description = """
Looking for a Data Scientist with skills in Python, Machine Learning,
Deep Learning, SQL, and experience with TensorFlow or PyTorch.
"""

# Strong Candidate
resume_strong = """
Skills: Python, Machine Learning, Deep Learning, SQL, TensorFlow
Experience: 3 years in Data Science
"""

# Average Candidate
resume_average = """
Skills: Python, Machine Learning, SQL
Experience: 1 year
"""

# Weak Candidate
resume_weak = """
Skills: Excel, Data Entry
Experience: Fresher
"""
```

```python
extract_prompt = PromptTemplate(
    input_variables=["resume"],
    template="""
Extract the following from the resume:
- Skills
- Experience
- Tools

Resume:
{resume}
"""
)
```

```python
score_prompt = PromptTemplate(
    input_variables=["resume", "job"],
    template="""
Score the candidate from 0 to 100 based on match.

Rules:
- Do NOT assume missing skills
- Only use given information

Resume:
{resume}

Job Description:
{job}
"""
)
```

```python
explain_prompt = PromptTemplate(
    input_variables=["resume", "job"],
    template="""
Explain why this candidate received the score.

Resume:
{resume}

Job Description:
{job}
"""
)
```

```python
import os
from langchain_openai import ChatOpenAI  # Updated import path

# Method 1: Set the API key in your environment variables
os.environ["OPENAI_API_KEY"] = "your-api-key-here"  # Replace with your actual OpenAI API key

# Method 2: Or pass the API key directly to the ChatOpenAI constructor
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    api_key="your-api-key-here"  # Replace with your actual OpenAI API key
)

# Now you can use the llm object for generating responses
```

```python
def extract_skills(resume):
    prompt = extract_prompt.format(resume=resume)
    return llm.invoke(prompt)
```

```python
def match_skills(resume, job):
    prompt = match_prompt.format(resume=resume, job=job)
    return llm.invoke(prompt)
```

```python
def score_candidate(resume, job):
    prompt = score_prompt.format(resume=resume, job=job)
    return llm.invoke(prompt)
```

```python
def run_pipeline(resume, job):
    print("---- Skill Extraction ----")
    print(extract_skills(resume))

    print("\n---- Matching ----")
    print(match_skills(resume, job))

    print("\n---- Score ----")
    print(score_candidate(resume, job))

    print("\n---- Explanation ----")
    print(explain_score(resume, job))
```

```python
print("===== STRONG CANDIDATE =====")
run_pipeline(resume_strong, job_description)

print("\n===== AVERAGE CANDIDATE =====")
run_pipeline(resume_average, job_description)

print("\n===== WEAK CANDIDATE =====")
run_pipeline(resume_weak, job_description)
print("\n===== DEBUG TEST =====")
resume_bad = "I enjoy cooking and music"
run_pipeline(resume_bad, job_description)
```
